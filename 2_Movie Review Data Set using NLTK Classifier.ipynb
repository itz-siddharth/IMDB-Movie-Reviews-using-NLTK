# Data Cleaning 

### Part of speech (pos_tag)

```python
from nltk import pos_tag
w = 'better'
#pos_tag(w) -->[('b', 'NN'), ('e', 'NN'), ('t', 'NN'), ('t', 'NN'), ('e', 'NN'), ('r', 'NN')]
pos_tag([w])
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

### Stopwords and Punctuations 

```python
from nltk.corpus import stopwords
import string

stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)

#stop = stop + punctuations
stops.update(punctuations)
stops
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            print(pos)
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
            
    print(output_words)
    return output_words  
```

```python
documents = [(clean_review(document), category) for document,category in documents]
```

```python
documents[0]
```

# Building Feature Set

#### NLTK classifiers require data to be in arrray of tuples. Each tuple consist 'dictionary of features' and 'category' 

#### choose top words and create feature set

```python
#75%
training_document = documents[:1500]

testing_document = documents[1500:]
#25%
```

```python
all_words = []

for doc in training_document:
    all_words += doc[0]
```

```python
import nltk
#Gives freq of each words
freq = nltk.FreqDist(all_words)
freq
```

```python
#3000 most common words
common = freq.most_common(3000)
common
```

```python
features = [i[0] for i in common]
features
```

```python
def get_features_dict(words):
    current_features = {}
    words_set = set(words)
    
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
training_data = [(get_features_dict(doc), category) for doc,category in training_document]
```

```python
training_data[0]
```

```python
testing_data = [(get_features_dict(doc), category) for doc,category in testing_document]
```

```python
testing_data[0]
```

# Classification Using NLTK Naive Bayes

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier, testing_data)
```
